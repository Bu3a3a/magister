```python
def draw_plot(df, style=PLOT_STYLE, labels=PLOT_LABELS):
    ax1 = df.plot(**style)
    ax1.set(**labels)
    #ax1.set_xticklabels(range(2007,2015))
    plt.legend(loc='best')
    return ax1
```

```python
def make_first_letters_shortcut(unistring, shortcut_length):
    alnum_ascii_str = ''.join(ch for ch in unidecode(unistring.lower()) if ch.isalnum() or ch == ' ')
    shortcut, counter = '', 0
    for word in alnum_ascii_str.split():
        if counter >= shortcut_length:
            break
        shortcut += word[0]
        counter += 1
    return shortcut
```

```python
lasso, ridge, ard = Lasso(), Ridge(), ARDRegression()
N = len(X)
idxs = np.arange(N)
cv = [(idxs[:i], idxs[i:i+1]) for i in range(1, N, 2)][3:]
print cv
```

```python
lasso.fit(X, Y.iloc[:,[0]])
ridge.fit(X, Y.iloc[:,[0]])
ard.fit(X, Y.iloc[:,[0]])
scores_1 = cross_val_score(lasso, X, Y.iloc[:,[0]], cv=cv, scoring='neg_mean_absolute_error')
scores_2 = cross_val_score(ridge, X, Y.iloc[:,[0]], cv=cv, scoring='neg_mean_absolute_error')
scores_3 = cross_val_score(ard, X, Y.iloc[:,[0]], cv=cv, scoring='neg_mean_absolute_error')
print u'Прогнозируемая переменная: ' + unicode(Y.columns[0]) + u'.\n';
print 'LASSO\n'
print 'Значения среднеквадратической ошибки для каждого тренировочного/тестового набора:\n'
print scores_1
print '\nСредняя величина среднеквадратической ошибки:\n'
print scores_1.mean()
print '\nКоэффициенты:\n'
print lasso.coef_
print '\n'
print 'RIDGE\n'
print 'Значения среднеквадратической ошибки для каждого тренировочного/тестового набора:\n'
print scores_2
print '\nСредняя величина среднеквадратической ошибки:\n'
print scores_2.mean()
print '\nКоэффициенты:\n'
print ridge.coef_
print '\n'
print 'ARD\n'
print 'Значения среднеквадратической ошибки для каждого тренировочного/тестового набора:\n'
print scores_3
print '\nСредняя величина среднеквадратической ошибки:\n'
print scores_3.mean()
print '\nКоэффициенты:\n'
print ard.coef_
print '\n'
```
