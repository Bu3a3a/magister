```python
def draw_plot(df, style=PLOT_STYLE, labels=PLOT_LABELS, xticks=range(2007,2015)):
    ax1 = df.plot(**style)
    ax1.set(**labels)
    ax1.set_xticklabels(xticks)
    plt.legend(loc='best')
    return ax1
```

```python
def make_first_letters_shortcut(unistring, shortcut_length):
    alnum_ascii_str = ''.join(ch for ch in unidecode(unistring.lower()) if ch.isalnum() or ch == ' ')
    shortcut, counter = '', 0
    for word in alnum_ascii_str.split():
        if counter >= shortcut_length:
            break
        shortcut += word[0]
        counter += 1
    return shortcut
```

```python
Y = pd.read_excel('new_targets.xlsx').transpose()
display(Y)
draw_plot(Y)
```

```python
X = pd.read_excel('new_data.xlsx').transpose()
display(X)
```
